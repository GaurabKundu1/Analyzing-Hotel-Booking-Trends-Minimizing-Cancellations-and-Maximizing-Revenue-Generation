```python
# change the date column to date and time format
# Here we will change the "reservation_status_date" because it is useful column and it is used as main for analysis
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], format='%d/%m/%Y')
```

```python
# Now we take out all the type of categories present 
for col in df.describe(include='object').columns:
    print(col)
    print(df[col].unique())
    print('-'*50)
```

```python
# check for any missing values
df.isnull().sum()
```

```python
'''
Here the missing value for the columns "company"and "agent" is so much ,it is around equal to 
total dataset so we wil remove the complete column
'''
df.drop(['agent','company'],axis=1, inplace=True)

'''
Now total dataset is 1 lakh around and removing 500 row of "country" column will not affect that much so
but filling all the values will be difficult so better will be removing that rows
'''
df.dropna(inplace=True)
```

```python
# Now we will check the missing Values again
df.isnull().sum()
```

```python
# lets check the Summary Statistics of Numerical columns
df.describe()
```

```python
# Removing outlier
df['adr'].plot(kind='box')
# the adr has a high outlier and we are going to use adr for analysis that why outlier removal was mandetory
df = df[df['adr']<5000]
```

### Data Analysis and Visualization

```python
# 1. check the amount of reservation which are canceled and which are not canceled
canceled_percentages = df['is_canceled'].value_counts(normalize=True)
'''
values_counts is the function which return the category name and number of times present
and to return the in the percentage form we will give "normalised = True"
'''
print(canceled_percentages)

plt.figure(figsize=(5,4))
plt.title('Reservation Status Count')
plt.bar(['Not Canceled', 'Canceled'],df['is_canceled'].value_counts(),edgecolor= 'k', width=0.7)
plt.show()
```

```python
# 2. Now we need to check which hotel has higest cancelation ratio
plt.figure(figsize=(8,4))
ax1 = sns.countplot(x = 'hotel', hue= 'is_canceled', data = df, palette= 'Blues')
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status in Different Hotels', size = 20)
plt.xlabel('Hotel')
plt.ylabel('Number of reservation')
plt.legend(['Not Canceled', 'Canceled'])
plt.show()
```

```python
#3. check the percentage of cancelation for both the hotel
#This is for the Resort hotel
resort_hotel=df[df['hotel']=='Resort Hotel'] #filter the data
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#This is for the City hotel
city_hotel=df[df['hotel']=='City Hotel'] #filter the data
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#4 Is there is affect of price in the cancellation of hotel
# Here we are gouping based on the reservation_data and mean price 
resort_hotel_group = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel_group=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title("Average Daily Rate on City and Resort Hotel",fontsize=30)
plt.plot(resort_hotel_group.index,resort_hotel_group['adr'],label="Resort Hotel")
plt.plot(city_hotel_group.index,city_hotel_group['adr'],label="City Hotel")
plt.legend(fontsize=20)
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month_name(locale = 'English')
df['year'] = df['reservation_status_date'].dt.year

# Create a list of months in the desired order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(16, 8))
ax1 = sns.countplot(x=df['month'], hue='is_canceled', data=df, palette='bright', order=month_order)
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1, 1))
plt.title("Reservation status per month", size=20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend(['Not Canceled', 'Canceled'])
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.title("ADR per month", fontsize=30)
df['month_year'] = df['month']
sns.barplot(x='month_year', y='adr', data=df[df['is_canceled'] == 1].groupby('month_year')[['adr']].sum().reset_index(), order=month_order)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.xlabel('Month')
plt.ylabel('ADR')
plt.show()
```

```python
#7 Let check the cancelation rate based on the countries
# we will see just the top 10 countries
cancelled_data=df[df['is_canceled'] == 1]
not_cancelled_data=df[df['is_canceled'] == 0]
top_10_countries = cancelled_data['country'].value_counts()[:10]

#from the graph we can find that PRT country has the higest cancellation 
plt.figure(figsize=(8,8))
plt.title('Top 10 Countries with reservation canceled')
plt.pie(top_10_countries, autopct='%.2f', labels=top_10_countries.index)

# Add a circle at the center of the pie chart to make it look like a donut chart
center_circle = plt.Circle((0, 0), 0.60, fc='white')
plt.gca().add_artist(center_circle)

plt.show()
```

```python
#10 let's check the price whethere the hotel which is cancelled is high or the hotel which is not canceld is high
# Calculate the average daily rate for cancelled bookings over different dates
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)  # Reset the index to ensure proper DataFrame structure
cancelled_df_adr.sort_values("reservation_status_date", inplace=True)  # Sort the DataFrame by date

# Calculate the average daily rate for not cancelled bookings over different dates
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)  # Reset the index for proper DataFrame structure
not_cancelled_df_adr.sort_values("reservation_status_date", inplace=True)  # Sort the DataFrame by date

# Create a plot to visualize the comparison of ADR between cancelled and not cancelled bookings over time
plt.figure(figsize=(20, 6))  # Set the figure size for the plot
plt.title("Average Daily Rate", fontsize=30)  # Set the title of the plot
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled')  # Plot ADR for not cancelled bookings
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled')  # Plot ADR for cancelled bookings
plt.legend(fontsize=20)  # Add legend with specified font size
plt.show()  # Display the plot
```
